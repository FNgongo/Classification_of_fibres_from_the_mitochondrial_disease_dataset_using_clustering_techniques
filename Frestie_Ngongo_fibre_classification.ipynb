## 6a K-means Clustering

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
from time import time
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

#Defining data 
df3 = df
df3 == df3
#Creating x and label object
X = df3[['logNDUFA13','logNDUFB8','logMTCO1', 'logCOX4+4L2', 'logOSCP', 'logSDHA', 'logUqCRC2', 'logVDAC1']]
labels = df3[['cell_id']]
#Making our category for k-mean as integer
labels.cell_id = labels.cell_id.astype("category")
labels.cell_id = labels.cell_id.cat.codes
labels.cell_id = labels.cell_id.astype("int")

#Fitting the k-mean model
cls = KMeans(n_clusters = 2)
lab = cls.fit(X)
lab.labels_ #Printing the label
```

```python
#K-means predicting who are patients and who are control
df3['patient_id'] = lab.labels_
print(df3['patient_id'])
```
